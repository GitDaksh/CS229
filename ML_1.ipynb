```python
import numpy as np
import pandas as pd

# machine learning
import keras
import ml_edu.experiment
import ml_edu.results

# data visualization
import plotly.express as px
```

```python
chicago_taxi_dataset = pd.read_csv("https://download.mlcc.google.com/mledu-datasets/chicago_taxi_train.csv")
```

```python
training_df = chicago_taxi_dataset.loc[:, ('TRIP_MILES', 'TRIP_SECONDS', 'FARE', 'COMPANY', 'PAYMENT_TYPE', 'TIP_RATE')]

print('Read dataset completed successfully.')

print('Total number of rows: {0}\n\n'.format(len(training_df.index)))

print(training_df.head(200))
```

```python
print('Total number of rows: {0}\n\n'.format(len(training_df. index)))

training_df.describe(include='all')
```

```python
#What is the maximum fare?
max_fare = training_df['FARE'].max()
print('The maximum fare is: ${:.2f}'.format(max_fare))

mean_distance = training_df['TRIP_MILES'].mean()
print('The mean distance of the whole trip is: {:.2f}'.format(mean_distance))

unique_cabs = training_df['COMPANY'].nunique()
print('Total cab companies: {:.2f}'.format(unique_cabs))

most_frequest_payment_type = training_df['PAYMENT_TYPE'].value_counts().idxmax()
print('Most frequent payment type: {}'.format(most_frequest_payment_type))

missing_values = training_df.isnull().sum().sum()
print("Are there any missing values: ", "NO" if missing_values == 0 else "YES")
```
